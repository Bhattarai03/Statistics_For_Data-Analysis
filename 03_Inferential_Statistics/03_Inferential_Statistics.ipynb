```python
#For datasets
#It is a population dataset
from random import randint
df=pd.DataFrame({
    "Sales":np.random.randint(100,1000,600)
})
print(df)
```

```python
#Random sample of 50
sample=df["Sales"].sample(50,random_state=42)
print(sample)
```

```python
#Point Estimation
population_mean=df["Sales"].mean()
sample_mean=sample.mean()
sample_mean,population_mean
```

```python
#Interval Estimation
>provide intervals in which population parameter falls
>More accurate than point estimation
```

```python
import scipy.stats as stats

n = len(sample)
std = sample.std()

# 95% CI
confidence_interval = stats.t.interval(
    0.95,       # confidence level
    n-1,        # degrees of freedom
    loc=sample_mean,   # sample mean
    scale=std/np.sqrt(n)  # standard error
)

confidence_interval
```

```python
#Chi _ Square test
import pandas as pd
from scipy.stats import chi2_contingency

# Creating a simple contingency table
data = {
    "Purchased": [40, 50],
    "Not_Purchased": [60, 50]
}
index = ["Male", "Female"]

df = pd.DataFrame(data, index=index)

print("Contingency Table:")
display(df)

# Performing Chi-Square Test of Independence
chi2, p_value, dof, expected = chi2_contingency(df)

print("\nChi-Square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p_value)

# Expected frequency table
print("\nExpected Frequencies:")
expected_df = pd.DataFrame(expected, index=index, columns=df.columns)
display(expected_df)

# Decision
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: Reject Null Hypothesis")
    print("➡ Gender and Purchase decision are NOT independent.")
else:
    print("\nConclusion: Fail to Reject Null Hypothesis")
    print("➡ Gender and Purchase decision are independent.")
```

```python
# Anova Testing
import numpy as np
from scipy.stats import f_oneway

# Scores from three teaching methods
method_A = np.array([85, 78, 92, 88, 76])
method_B = np.array([72, 70, 68, 65, 74])
method_C = np.array([90, 88, 85, 87, 92])

print("Method A Scores:", method_A)
print("Method B Scores:", method_B)
print("Method C Scores:", method_C)

# One-way ANOVA test
f_statistic, p_value = f_oneway(method_A, method_B, method_C)

print("\nF-statistic:", f_statistic)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: Reject Null Hypothesis")
    print("➡ At least one teaching method has a significantly different mean score.")
else:
    print("\nConclusion: Fail to Reject Null Hypothesis")
    print("➡ All teaching methods have similar mean scores (no significant difference).")
```

```python
import numpy as np
from scipy.stats import f_oneway

# Scores from three teaching methods
method_A = np.array([85, 78, 92, 88, 76])
method_B = np.array([72, 70, 68, 65, 74])
method_C = np.array([90, 88, 85, 87, 92])

print("Method A Scores:", method_A)
print("Method B Scores:", method_B)
print("Method C Scores:", method_C)

# One-way ANOVA test
f_statistic, p_value = f_oneway(method_A, method_B, method_C)

print("\nF-statistic:", f_statistic)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: Reject Null Hypothesis")
    print("➡ At least one teaching method has a significantly different mean score.")
else:
    print("\nConclusion: Fail to Reject Null Hypothesis")
    print("➡ All teaching methods have similar mean scores (no significant difference).")
```

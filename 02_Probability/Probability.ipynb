```python
## ðŸ“˜ 1. Probability Distribution 

A probability distribution describes how probabilities are assigned to different possible values of a random variable.
It shows:
>What values the variable can take
>How likely each value is

### ðŸ“Œ Types of Probability Distributions
A. Discrete Probability Distribution
Used when the variable takes countable values.
Examples:
Number of customers entering a shop
Number of defects in 50 products
Flipping a coin

#Common discrete distributions:
>Binomial Distribution â€“ number of successes in fixed trials
>Poisson Distribution â€“ number of events in a fixed time/space
>Geometric Distribution â€“ trials until the first success

B. Continuous Probability Distribution
Used when the variable takes infinite real values.
Examples:
Height
Weight
Salary
Temperature

#Common continuous distributions:
>Normal Distribution â€“ most natural data follows a bell-shaped curve
>Uniform Distribution â€“ equal probability for all values
>Exponential Distribution â€“ time between events

ðŸ“Œ Why Probability Distributions Matter
> data patterns
>Select correct statistical tests
>Model uncertainty
>Analyze behavior of random variables
>Assist in real-world decision making

## ðŸ“˜ 2. Estimation 
Estimation is the process of using sample data to estimate population parameters.
ðŸ“Œ A. Point Estimation
Produces a single value to estimate a population parameter.
Examples:
>Sample mean estimates population mean
>Sample variance estimates population variance
ðŸ“Œ B. Interval Estimation (Confidence Intervals)
Provides a range within which the true population parameter is likely to lie.
Example:
â€œMean salary is 45k Â± 3k at 95% confidence level.â€

ðŸ“Œ Why Estimation is Important
>Measures sampling uncertainty
>Used in hypothesis testing
>Important for forecasting
>Helps in decision-making with incomplete information

## ðŸ“˜ 3. Conditional Probability 
Conditional probability measures the probability of an event given that another event has already occurred.
Formula: P(A|B)=p(AnB)/P(B)
ðŸ“Œ Example
>Probability a customer buys a phone given they clicked the product page.
>Probability of rain given that humidity is high.
ðŸ“Œ Uses
>Customer behavior analysis
>Recommendation systems
>Medical diagnosis
>Risk analysis

## ðŸ“˜ 4. Bayes Probability (Bayes Theorem)
Bayes theorem allows updating an earlier belief (prior probability) when new information becomes available.
Formula: P(A|B)=(P(B|A) *P(A))/P(B)

 ðŸ“Œ Terminology
P(A) â†’ Prior probability
P(B|A) â†’ Likelihood
P(A|B) â†’ Posterior probability (updated probability)
P(B) â†’ Evidence

 ðŸ“Œ Real-Life Use Cases
>Spam detection
>Medical test accuracy
>Fraud prediction
>Weather forecasting
>Naive Bayes classifier (Machine Learning)
```

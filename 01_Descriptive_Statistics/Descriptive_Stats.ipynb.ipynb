```python
"Lets clean the above dataset"
```

```python
df["Age"].fillna(df["Age"].median())
```

```python
df["Salary"].fillna(df["Salary"].median())
```

```python
df["JoinYear"].fillna(df["JoinYear"].median())
```

```python
print(df)
```

```python
df["Gender"].replace({'M':'Male' , 'F':'Female'})
```

```python
df["Department"]=df["Department"].str.capitalize()
```

```python
print(df)
```

```python
df['Satisfaction'].fillna(df['Satisfaction'].median())
```

```python
print(df)
```

```python
Q1=df["Salary"].quantile(0.25)
Q3=df["Salary"].quantile(0.75)
IQR=Q3-Q1
df=df[(df["Salary"]>=(Q1-IQR)) & (df["Salary"]<=(Q3+IQR))]
```

```python
print(df)
```

```python
df['Age'] = df['Age'].astype(int)
df['Salary'] = df['Salary'].astype(int)
df['JoinYear'] = df['JoinYear'].astype(int)
df['Satisfaction'] = df['Satisfaction'].astype(int)

df['Gender'] = df['Gender'].astype('category')
df['Department'] = df['Department'].astype('category')
```

```python
df.to_csv("Clean_Descriptive_Dataset")
```
